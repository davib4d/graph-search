```python
# Grafo de exemplo (Lista de Adjacência)
grafo = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E', 'F'],
    'C': ['A', 'G', 'H'],
    'D': ['A', 'I', 'J'],
    'E': ['B', 'K', 'L'],
    'F': ['B', 'M'],
    'G': ['C', 'N'],
    'H': ['C', 'O'],
    'I': ['D', 'P'],
    'J': ['D', 'K'],
    'K': ['E', 'J', 'P'],
    'L': ['E', 'M'],
    'M': ['F', 'L', 'N'],
    'N': ['G', 'M'],
    'O': ['H', 'P'],
    'P': ['I', 'K', 'O']
}

print(f"Grafo carregado com {len(grafo)} vértices.")
```

```python
def dfs(grafo, inicio):
    visitados = set()
    pilha = [inicio]
    ordem_visita = []

    while pilha:
        vertice = pilha.pop() # Remove do final (LIFO)

        if vertice not in visitados:
            visitados.add(vertice)
            ordem_visita.append(vertice)
            
            # Adiciona vizinhos à pilha
            # Invertemos a lista apenas para manter a ordem alfabética "intuitiva" na saída, não é obrigatório
            for vizinho in reversed(grafo[vertice]):
                if vizinho not in visitados:
                    pilha.append(vizinho)
    
    return ordem_visita
```

```python
# Testando DFS começando do nó 'A'
resultado_dfs = dfs(grafo, 'A')
print(f"Ordem de visita DFS: {resultado_dfs}")
# Esperado algo como: ['A', 'B', 'D', 'E', 'F', 'C'] (vai o mais fundo possível antes de voltar)
```
