```python
# Grafo de exemplo (Lista de Adjacência)
grafo = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E', 'F'],
    'C': ['A', 'G', 'H'],
    'D': ['A', 'I', 'J'],
    'E': ['B', 'K', 'L'],
    'F': ['B', 'M'],
    'G': ['C', 'N'],
    'H': ['C', 'O'],
    'I': ['D', 'P'],
    'J': ['D', 'K'],
    'K': ['E', 'J', 'P'],
    'L': ['E', 'M'],
    'M': ['F', 'L', 'N'],
    'N': ['G', 'M'],
    'O': ['H', 'P'],
    'P': ['I', 'K', 'O']
}

print(f"Grafo carregado com {len(grafo)} vértices.")
```

```python
from collections import deque

def bfs(grafo, inicio):
    visitados = set()
    fila = deque([inicio])
    ordem_visita = []

    visitados.add(inicio)

    while fila:
        vertice = fila.popleft() # O(1) - Remove do início (FIFO)
        ordem_visita.append(vertice)

        for vizinho in grafo[vertice]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)
    
    return ordem_visita
```

```python
# Testando BFS começando do nó 'A'
resultado_bfs = bfs(grafo, 'A')
print(f"Ordem de visita BFS: {resultado_bfs}")
# Esperado algo como: ['A', 'B', 'C', 'D', 'E', 'F'] (explora vizinhos diretos primeiro)
```
